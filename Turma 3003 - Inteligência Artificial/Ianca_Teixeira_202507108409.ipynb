```python
import pandas as pd;
```

```python
dados = pd.read_excel("base_dados_ia.xlsx");
```

```python
print(dados.head());
# Impressão dos primeiros dados
```

```python
dados["media"] = dados[["nota1", "nota2"]].mean(axis=1);
# Passo de calculo de media de cada aluno.
```

```python
dados["aprovaçao"] = (dados["media"] >= 6) & (dados["frequencia"] >= 75);
dados["reprovado"] = (dados["media"] < 6) & (dados["frequencia"] < 75);
#Definição de aprovação, onde a media deve ser maior ou igual a 6.
```

```python
print(dados[["nome", "media", "frequencia", "aprovaçao"]]);
# Impressão dos dados com a media e aprovação.
```
